```python
import re

import gensim
from gensim.models import Word2Vec
```

Get the text of *Moby Dick*.

```python
!wget https://raw.githubusercontent.com/jonathanblaney/CC1-Michaelmas25/refs/heads/main/week6-AI/moby-dick.txt
```

```python
with open('moby-dick.txt', 'r') as f:
  text = f.read()
```

Delete new lines.

```python
text = text.replace("\n", " ")
```

Normalise the text as lower case and delete punctuation.

```python
text = text.lower()
text = re.sub(r"[%&'()*+—,./:;?_’“”-]", "", text)
```

Split the text on spaces into a list.

```python
words = text.split()
```

Word2Vec expects multiple documents, so to do *Moby Dick* on its own we'll split it into two lists. It would make more sense to split it into chapters, but that would take more data work.

```python
len(words)
```

```python
moby1 = words[:100_000]
moby2 = words[100_001:]
```

```python
moby_all = []
moby_all.append(moby1)
moby_all.append(moby2)
```

```python
# train the model
model = Word2Vec(sentences=moby_all, window=5, min_count=3, workers=4, epochs=5, sg=1)

# save the model
model.save("word2vec.model")
```

```python
model = Word2Vec.load("word2vec.model")
```

Now we can query the word embeddings of *Moby Dick*.

```python
model.wv.most_similar('whaling', topn=20)
```

```python
model.wv.most_similar('ahab', topn=20)
```

```python
model.wv.most_similar('ship', topn=20)
```
